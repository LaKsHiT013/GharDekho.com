```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Datasets/merged(flats_&_houses)_CLEANED2(Missing_Values_Treated)')
```

```python
df.shape
```

```python
df.head()
```

```python
train_df = df.drop(columns=['society','price_per_sqft'])
```

```python
train_df.head()
```

```python
numeric_train_df = train_df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_train_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
numeric_train_df = train_df.select_dtypes(include=['float64', 'int64'])
correlation_with_price = numeric_train_df.corr()['price'].sort_values(ascending=False)
print(correlation_with_price)
```
